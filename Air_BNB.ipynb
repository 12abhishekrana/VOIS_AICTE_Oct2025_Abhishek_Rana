```python
# Drop duplicate records
df.drop_duplicates(inplace=True)

# Drop 'house_rules' and 'license' columns if they exist
df.drop(['house_rules', 'license'], axis=1, inplace=True, errors='ignore')

# Remove all dollar signs in the price and service fee columns (if they exist)
if 'price' in df.columns:
    df['price'] = df['price'].str.replace('$', '', regex=False)
if 'service fee' in df.columns:
    df['service fee'] = df['service fee'].str.replace('$', '', regex=False)

# Remove all commas in the price and service fee columns (if they exist)
if 'price' in df.columns:
    df['price'] = df['price'].str.replace(',', '', regex=False)
if 'service fee' in df.columns:
    df['service fee'] = df['service fee'].str.replace(',', '', regex=False)

# Rename the price and service fee columns to include a dollar sign
df.rename(columns={'price': 'price_$', 'service fee': 'service_fee_$'}, inplace=True)

# Drop all records with missing values
df.dropna(inplace=True)

# Change all mismatched data types to the appropriate ones (if columns exist)
if 'price_$' in df.columns:
    df['price_$'] = df['price_$'].astype(float)
if 'service_fee_$' in df.columns:
    df['service_fee_$'] = df['service_fee_$'].astype(float)
if 'id' in df.columns:
    df['id'] = df['id'].astype(str)
if 'host id' in df.columns:
    df['host id'] = df['host id'].astype(str)
if 'last review' in df.columns:
    df['last review'] = pd.to_datetime(df['last review'], errors='coerce')
if 'Construction year' in df.columns:
    df['Construction year'] = pd.to_numeric(df['Construction year'], errors='coerce').astype('Int64')

# Correct the spelling of 'brookin' to 'Brooklyn' (if column exists)
if 'neighbourhood group' in df.columns:
    df.loc[df['neighbourhood group'].str.lower() == 'brookin', 'neighbourhood group'] = 'Brooklyn'

# Get rid of outliers in the 'availability 365' column data (if column exists)
if 'availability 365' in df.columns:
    df.drop(df[df['availability 365'] > 506].index, inplace=True)
```

```python
property_types = df['room type'].value_counts().to_frame()
property_types

```

```python
room_type_bar = plt.bar(property_types.index, property_types.loc[:,"count"]);

# Add labels to bars
plt.bar_label(room_type_bar, labels=property_types.loc[:,"count"], padding=4);

# Set limits and labels
plt.ylim([0, 50000]);
plt.xlabel('Room Type');
plt.ylabel('Room Type Count');
plt.title('Property Types and their count in the Dataset');

# Show the plot
plt.show()
```

```python
hood_group = df['neighbourhood group'].value_counts().to_frame()
hood_group
```

```python
# Which neighborhood group has the highest number of listings?
hood_group = df['neighbourhood group'].value_counts().to_frame(name="count")

# Bar chart
hood_group_bar = plt.bar(hood_group.index, hood_group["count"])

# Add labels to bars
plt.bar_label(hood_group_bar, labels=hood_group["count"], padding=4)

# Chart styling
plt.ylim([0, 40000])
plt.xlabel('Neighborhood Groups')
plt.ylabel('Number of Listings')
plt.xticks(rotation=45)
plt.title('Which Neighborhood Group has the Highest Number of Listings')

# Show plot
plt.show()
```

```python
# Which neighborhood groups have the highest average prices for Airbnb listings?
avg_price = df.groupby('neighbourhood group')['price_$'].mean().sort_values(ascending=False).to_frame()

# Bar chart
avg_price_bar = plt.bar(avg_price.index, avg_price['price_$'])

# Add labels to bars
plt.bar_label(avg_price_bar, labels=round(avg_price['price_$'], 2), label_type='edge', padding=4)

# Chart styling
plt.ylim([0, 700])
plt.xlabel('Neighborhood Group')
plt.ylabel('Average Price per Listing ($)')
plt.xticks(rotation=45)
plt.title('Average Price per Listing ($) in Each Neighborhood Group')

# Show plot
plt.show()
```

```python
# Relationship between construction year and price
avg_price_by_year = df.groupby('Construction year')['price_$'].mean().to_frame()

# Plot
avg_price_by_year.plot(figsize=(10, 5), legend=False, color='royalblue', linewidth=2)

plt.xlabel('Construction Year')
plt.ylabel('Average Price ($)')
plt.title('Average Price ($) for Properties by Construction Year')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python

# Top 10 hosts by calculated host listing count
hosts = (
    df.groupby('host name')['calculated host listings count']
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .to_frame()
)

# Plot
hosts_bar = plt.bar(hosts.index, hosts['calculated host listings count'], color='teal')
plt.bar_label(hosts_bar, labels=hosts['calculated host listings count'], padding=3)
plt.xlabel('Host Name')
plt.ylabel('Calculated Host Listings Count')
plt.xticks(rotation=80)
plt.ylim([0, 120000])
plt.title('Top 10 Hosts by Calculated Host Listings Count')
plt.tight_layout()
plt.show()
```

```python
# Are hosts with verified identities more likely to receive positive reviews?
review = df.groupby('host_identity_verified')['review rate number'].mean().sort_values(ascending=False).to_frame()
# Display the result
review
```

```python

# Bar chart for average review rate by host verification status
review_bar = plt.bar(review.index, review['review rate number'], color='coral')
plt.bar_label(review_bar, labels=round(review['review rate number'], 2), padding=4)
plt.ylim([0, 4])
plt.xlabel('Host Verification Status')
plt.ylabel('Average Review Rate Number')
plt.title('Average Review Rate for Each Verification Status')
plt.show()
```

```python
# Check correlation between price and service fee
correlation = df['price_$'].corr(df['service_fee_$'])
print("Correlation between Price and Service Fee:", correlation)
```

```python
# Plot regression line to visualize the relationship
sns.regplot(data=df, x='price_$', y='service_fee_$', scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.xlabel('Price ($)')
plt.ylabel('Service Fee ($)')
plt.title('Regression Plot: Correlation between Price and Service Fee')
plt.show()
```

```python
# Average Review Rate Number (ARRN) by Neighborhood Group and Room Type
ARRN = (
    df.groupby(['neighbourhood group', 'room type'])['review rate number']
    .mean()
    .round(2)
    .to_frame(name='Average Review Rate Number')
)

# Display the result
ARRN
```

```python
plt.figure(figsize=(12, 10))
sns.barplot(data=df, x='neighbourhood group', y='review rate number', hue='room type', errorbar=None)

plt.xlabel('Neighbourhood Group')
plt.ylabel('Average Review Rate')
plt.title('Average Review Rate for Each Room/Property Type in Each Neighbourhood Group')
plt.legend(title='Room Type')
plt.show()
```

```python
# Regression plot showing the relationship between host listings count and availability
sns.regplot(
    data=df,
    x='calculated host listings count',
    y='availability 365',
    scatter_kws={'alpha': 0.5},
    line_kws={'color': 'red'}
)

plt.xlabel('Calculated Host Listings Count')
plt.ylabel('Availability (Days per Year)')
plt.title('Relationship Between Host Listing Count and Availability Throughout the Year')
plt.show()
```
